```python
# process state and carbon emissions data
import pandas as pd

file_path = '/Users/richardchu/Desktop/CS 416/NarrativeVisualization_Codes/jueichechu.github.io/data_and_preprocessing/carbon_dioxide_emission_by_year.xlsx'
df = pd.read_excel(file_path, header=None)

state_column = df.iloc[5:56, 0].reset_index(drop=True)
year_2021_column = df.iloc[5:56, 52].reset_index(drop=True)

# Create a new DataFrame with the extracted columns
new_df = pd.DataFrame({
    'State': state_column,
    'Carbon Dioxide Emissions (Million Metric Tons)': year_2021_column
})

csv_output_path = '/Users/richardchu/Desktop/CS 416/NarrativeVisualization_Codes/jueichechu.github.io/data_and_preprocessing/data.csv'
new_df.to_csv(csv_output_path, index=False)
```

```python
# process carbon emissions by fuel
file_path_fuel = '/Users/richardchu/Desktop/CS 416/NarrativeVisualization_Codes/jueichechu.github.io/data_and_preprocessing/carbon_dioxide_emission_by_fuel.xlsx'

df_data = pd.read_csv(csv_output_path)
df_fuel = pd.read_excel(file_path_fuel, header=None)

coal_column = df_fuel.iloc[4:55, 1].reset_index(drop=True)
petroleum_column = df_fuel.iloc[4:55, 2].reset_index(drop=True) 
natural_gas_column = df_fuel.iloc[4:55, 3].reset_index(drop=True)

df_data['Coal (MMT)'] = coal_column
df_data['Petroleum (MMT)'] = petroleum_column
df_data['Natural Gas (MMT)'] = natural_gas_column

df_data.to_csv(csv_output_path, index=False)
```

```python
# process population data
csv_output_path = '/Users/richardchu/Desktop/CS 416/NarrativeVisualization_Codes/jueichechu.github.io/data_and_preprocessing/data.csv'
pop_file_path = '/Users/richardchu/Desktop/CS 416/NarrativeVisualization_Codes/jueichechu.github.io/data_and_preprocessing/NST-EST2023-POP.xlsx'

df_data = pd.read_csv(csv_output_path)
df_pop = pd.read_excel(pop_file_path, header=None)

state_pop = df_pop.iloc[9:62, [0, 3]].dropna().reset_index(drop=True)
state_pop.columns = ['State', 'State Population']

state_pop['State'] = state_pop['State'].str.strip().str.lstrip('.')
df_data['State'] = df_data['State'].str.strip()

state_pop_dict = state_pop.set_index('State')['State Population'].to_dict()
df_data['State Population'] = df_data['State'].map(state_pop_dict)

df_data.to_csv(csv_output_path, index=False)
```

```python
# process income data

csv_output_path = '/Users/richardchu/Desktop/CS 416/NarrativeVisualization_Codes/jueichechu.github.io/data_and_preprocessing/data.csv'
income_file_path = '/Users/richardchu/Desktop/CS 416/NarrativeVisualization_Codes/jueichechu.github.io/data_and_preprocessing/median_family_income_by_state.xlsx'

df_data = pd.read_csv(csv_output_path)
df_income = pd.read_excel(income_file_path, header=None, skiprows=1, usecols=[0, 1], nrows=52)

df_income.columns = ['State', 'Median Family Income (U.S. Dollars)']
df_income['State'] = df_income['State'].str.strip()
df_data['State'] = df_data['State'].str.strip()

income_dict = df_income.set_index('State')['Median Family Income (U.S. Dollars)'].to_dict()
df_data['Median Family Income (U.S. Dollars)'] = df_data['State'].map(income_dict)

df_data.to_csv(csv_output_path, index=False)
```

```python
# process vehicle registration counts data
file_path = '/Users/richardchu/Desktop/CS 416/NarrativeVisualization_Codes/jueichechu.github.io/data_and_preprocessing/2021 Light-Duty Vehicle Registration Counts by State and Fuel Type.xlsx'

# Load the data and skip the first row to use the actual headers
df = pd.read_excel(file_path, skiprows=1)
df.columns = df.columns.str.strip()

df['Vehicle Registration Counts'] = df.loc[:, 'Electric (EV)':'Unknown Fuel'].sum(axis=1)

csv_output_path = '/Users/richardchu/Desktop/CS 416/NarrativeVisualization_Codes/jueichechu.github.io/data_and_preprocessing/vehicle_registration_counts.csv'
df.to_csv(csv_output_path, index=False)


csv_output_path = '/Users/richardchu/Desktop/CS 416/NarrativeVisualization_Codes/jueichechu.github.io/data_and_preprocessing/data.csv'
vehicle_file_path = '/Users/richardchu/Desktop/CS 416/NarrativeVisualization_Codes/jueichechu.github.io/data_and_preprocessing/vehicle_registration_counts.csv'

df_data = pd.read_csv(csv_output_path)
df_vehicle = pd.read_csv(vehicle_file_path)

state_col = df_vehicle.iloc[1:52, 0].reset_index(drop=True).str.strip()
value_col = df_vehicle.iloc[1:52, 13].reset_index(drop=True) 

df_vehicle_counts = pd.DataFrame({
    'State': state_col,
    'Vehicle Registration Counts': value_col
})

vehicle_counts_dict = df_vehicle_counts.set_index('State')['Vehicle Registration Counts'].to_dict()
df_data['Vehicle Registration Counts'] = df_data['State'].map(vehicle_counts_dict)
df_data.to_csv(csv_output_path, index=False)
```

```python
# final data processing for decimal places

csv_output_path = '/Users/richardchu/Desktop/CS 416/NarrativeVisualization_Codes/jueichechu.github.io/data_and_preprocessing/data.csv'
df_data = pd.read_csv(csv_output_path)

columns_one_decimal = ['Carbon Dioxide Emissions (Million Metric Tons)', 'Coal (MMT)', 'Petroleum (MMT)', 'Natural Gas (MMT)']
columns_no_decimal = ['State Population', 'Vehicle Registration Counts']

# round values to 1 decimal place
for col in columns_one_decimal:
    if col in df_data.columns:
        df_data[col] = df_data[col].apply(lambda x: f"{float(x):.1f}" if pd.notnull(x) and x != '' else "")

# round values to 0 decimal place (integer)
for col in columns_no_decimal:
    if col in df_data.columns:
        df_data[col] = df_data[col].apply(lambda x: f"{int(float(x))}" if pd.notnull(x) and x != '' else "")

df_data.to_csv(csv_output_path, index=False)
```

```python
# assign region to states

csv_output_path = '/Users/richardchu/Desktop/CS 416/NarrativeVisualization_Codes/jueichechu.github.io/data_and_preprocessing/data.csv'
df_data = pd.read_csv(csv_output_path)

regions = {
    "New England": ["Connecticut", "Maine", "Massachusetts", "New Hampshire", "Rhode Island", "Vermont"],
    "Middle Atlantic": ["New Jersey", "New York", "Pennsylvania"],
    "East North Central": ["Illinois", "Indiana", "Michigan", "Ohio", "Wisconsin"],
    "West North Central": ["Iowa", "Kansas", "Minnesota", "Missouri", "Nebraska", "North Dakota", "South Dakota"],
    "South Atlantic": ["Delaware", "District of Columbia", "Florida", "Georgia", "Maryland", "North Carolina", "South Carolina", "Virginia", "West Virginia"],
    "East South Central": ["Alabama", "Kentucky", "Mississippi", "Tennessee"],
    "West South Central": ["Arkansas", "Louisiana", "Oklahoma", "Texas"],
    "Mountain": ["Arizona", "Colorado", "Idaho", "Montana", "Nevada", "New Mexico", "Utah", "Wyoming"],
    "Pacific": ["Alaska", "California", "Hawaii", "Oregon", "Washington"]
}

def get_region(state):
    for region, states in regions.items():
        if state in states:
            return region
    return "Unknown"

df_data['Region'] = df_data['State'].apply(get_region)

df_data.to_csv(csv_output_path, index=False)
```
